```python
corates = [0.5, 0.65, 0.8]
muscales = [0.01, 0.025, 0.5]
popsizes = [100, 200, 350]
popmeths = ["pca"]

run_configs = {}
i = 0
# varying corates
muscale = 0.05
for popmeth in popmeths:
    for popsize in popsizes:
        for corate in corates:
            config = {
                "NUMBER_OF_ITERATIONS": 100,
                "NUMBER_OF_PATHS": 1,
                "PROJECTION_METHOD": popmeth,
                "CORPUS_METHOD": "read",
                "READ_CORPUS_SOURCE": "checkpoint",
                "SAMPLE_RATE": 22050,
                "POPULATION_SIZE": popsize,
                "CORPUS_SIZE": 40000,
                "MUTATION_STRENGTH": muscale,
                "MUTATION_RATE": 0.05,
                "CROSSOVER_RATE": corate,
                "ELITIST_SELECTION": "false",
                "UNIQUE_CLOSEST_IN_CORPUS": "false",
                "AUGMENT_RATIO_CROSSOVER": 0,
            }
            i += 1
            run_configs[str(i)] = config
# varying muscales
corate = 0.65
for popmeth in popmeths:
    for popsize in popsizes:
        for muscale in muscales:
            config = {
                "NUMBER_OF_ITERATIONS": 100,
                "NUMBER_OF_PATHS": 1,
                "PROJECTION_METHOD": popmeth,
                "CORPUS_METHOD": "read",
                "READ_CORPUS_SOURCE": "checkpoint",
                "SAMPLE_RATE": 22050,
                "POPULATION_SIZE": popsize,
                "CORPUS_SIZE": 40000,
                "MUTATION_STRENGTH": muscale,
                "MUTATION_RATE": 0.05,
                "CROSSOVER_RATE": corate,
                "ELITIST_SELECTION": "false",
                "UNIQUE_CLOSEST_IN_CORPUS": "false",
                "AUGMENT_RATIO_CROSSOVER": 0,
            }
            i += 1
            run_configs[str(i)] = config

run_configs
```
